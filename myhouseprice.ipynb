```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn import ensemble
from sklearn.metrics import mean_absolute_error
from sklearn.externals import joblib
from sklearn import preprocessing
```

```python
df =pd.read_csv("/home/nikhil/Desktop/Predicting-House-Prices-In-Bengaluru-Train-Data.csv")
```

```python
df.head()
```

```python
del df['availability']
del df['society']
```

```python
df.head()
```

```python
df['bath'] = df['bath'].fillna(2.0)
```

```python
df['balcony'] = df['balcony'].fillna(1)
```

```python
df = df.dropna(how='any')
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.head()
```

```python
features_df = pd.get_dummies(df, columns=['location','size','area_type'])
```

```python
features_df
```

```python
features_df.shape
```

```python
# Remove the sale price from the feature data
del features_df['price']
```

```python
# Create the X and y arrays
X = features_df.as_matrix()
y = df['price'].as_matrix()
```

```python
# Split the data set in a training set (70%) and a test set (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# Fit regression model
model = ensemble.GradientBoostingRegressor(
    n_estimators=1000,
    learning_rate=0.1,
    max_depth=6,
    min_samples_leaf=9,
    max_features=0.1,
    loss='huber',
    random_state=0
)
```

```python
model.fit(X_train, y_train)
```

```python
# Save the trained model to a file so we can use it in other programs
joblib.dump(model, 'myhouseprice.pkl')
```

```python
# Find the error rate on the training set
mse = mean_absolute_error(y_train, model.predict(X_train))
print("Training Set Mean Absolute Error: %.4f" % mse)
```

```python
# Find the error rate on the test set
mse = mean_absolute_error(y_test, model.predict(X_test))
print("Test Set Mean Absolute Error: %.4f" % mse)
```
